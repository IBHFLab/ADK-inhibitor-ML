```python
from matplotlib import font_manager, pyplot as plt

# 查找系统中所有可用的Times New Roman字体的路径
times_new_roman = font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
# 从列表中找到一个指定的字体名称，并确保字体能支持加粗
t_nr_path = [f for f in times_new_roman if 'Times New Roman' in f and 'Bold' in f]

if t_nr_path:
    # 如果找到了Times New Roman Bold字体，设置为默认字体
    prop = font_manager.FontProperties(fname=t_nr_path[0])
    plt.rcParams['font.family'] = prop.get_name()
    plt.rcParams['font.weight'] = 'bold'  # 设置字体为加粗
else:
    # 如果没有找到加粗的Times New Roman，尝试设置为普通的Times New Roman并加粗
    t_nr_path = [f for f in times_new_roman if 'Times New Roman' in f]
    if t_nr_path:
        prop = font_manager.FontProperties(fname=t_nr_path[0])
        plt.rcParams['font.family'] = prop.get_name()
        plt.rcParams['font.weight'] = 'bold'
    else:
        # 如果没有找到Times New Roman，使用默认的衬线字体并设置为加粗
        plt.rcParams['font.family'] = 'serif'
        plt.rcParams['font.weight'] = 'bold'
```

ECFPs-RF

```python
from y_scramble import Scrambler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score
import pandas as pd
import joblib
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 加载数据集
data = pd.read_csv('test-dataset-ECFPs.csv')
X_test = data.iloc[:, 2:].values  # 获取特征
Y_test = data.iloc[:, 1:2].values.ravel()  # 获取标签，并扁平化为一维数组

# 加载已训练的模型
model = joblib.load("RandomForest-ECFPs.pkl")

# 计算原始模型的AUC-ROC、准确率（accuracy）、精确率（precision）
Y_pred_prob = model.predict_proba(X_test)[:, 1]  # 获取预测的概率值（正类的概率）
Y_pred = model.predict(X_test)  # 获取模型的类别预测

# 计算原始模型的 AUC-ROC
original_auc = roc_auc_score(Y_test, Y_pred_prob)
# 计算准确率
original_acc = accuracy_score(Y_test, Y_pred)
# 计算精确率
original_prec = precision_score(Y_test, Y_pred)

# 输出原始模型的 AUC-ROC、准确率和精确率
print(f"Original model AUC-ROC: {original_auc:.3f}")
print(f"Original model Accuracy: {original_acc:.3f}")
print(f"Original model Precision: {original_prec:.3f}")

# 保存扰动结果，排除原始模型的第一行
def scramble_and_validate(model, X, Y, dataset_name="test"):
    """对模型进行扰动并保存结果"""
    scrambler = Scrambler(model=model, iterations=500)  # 设置扰动次数
    results = scrambler.validate(X, Y, scoring="roc_auc",  # 使用 AUC-ROC 作为评分指标
                                 cross_val_score_aggregator="mean", 
                                 pvalue_threshold=0.05, 
                                 as_df=True)
    # 排除第一行（原始模型结果），只保存扰动后的结果
    results_without_original = results.iloc[1:]
    # 保存扰动结果到CSV文件
    results_filename = f"{dataset_name}_validation_results.csv"
    results_without_original.to_csv(results_filename, index=False)
    print(f"Results saved to {results_filename}")
    return results_without_original

# 对测试集进行扰动验证并保存结果
test_results = scramble_and_validate(model, X_test, Y_test, "test")

# 获取扰动后的AUC-ROC分数
randomized_aucs = test_results['roc_auc'].values

# 计算扰动后AUC的显著性
t_stat, p_value = stats.ttest_1samp(randomized_aucs, original_auc)
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# 输出显著性分析结果
if p_value < 0.05:
    print("显著")
else:
    print("不显著")

import pandas as pd
import matplotlib.pyplot as plt

# 读取 test_validation_results.csv 文件
results = pd.read_csv('test_validation_results.csv')

# 提取 roc_auc 列
randomized_aucs = results['roc_auc'].values

# 绘制 AUC-ROC 的分布图
plt.figure(figsize=(7, 5))
plt.rc('axes', linewidth=1.7)  #线宽
# 设置x轴范围为 0 到 1，并且刻度间隔为 0.1
plt.xlim(0, 1)  # 设置x轴范围
plt.xticks(np.arange(0, 1.1, 0.1))  # 设置x轴刻度间隔为0.1
plt.hist(randomized_aucs, bins=40, alpha=1.0, label='Scrambled AUC-ROC',color='#40A578')
plt.axvline(original_auc, color='#604CC3', linestyle='dashed', linewidth=2, label=f'Original AUC-ROC: {original_auc:.3f}')
plt.xlabel('AUC-ROC', fontsize=15, weight='bold')
plt.ylabel('Frequency', fontsize=15, weight='bold')
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.title('RF-ECFPs', fontsize=17, weight='bold')
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(loc='upper right',frameon=True)

# 调整图像布局
plt.tight_layout()
plt.savefig('随机化 RF-ECFPs.svg', format='svg')
plt.show()
```
